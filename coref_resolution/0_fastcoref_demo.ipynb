## Quick start with an Example

```python
from fastcoref import FCoref

model = FCoref(device='cpu') # or FCoref(device='cuda:0') if you want to use a GPU

# read the text file (fulltext of ATel no. 16146)
with open('text_example') as f:
    f = f.read()
    
preds = model.predict(texts=[f])
```

```python
# get the clusters of mentions
preds[0].get_clusters()
```

```python
# get start and end spans of detected mentions
preds[0].get_clusters(as_strings=False)
```
